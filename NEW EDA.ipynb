# Age EDA 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Determining the range for age bins
min_age = final_prostate_data['AGEP_A'].min()
max_age = final_prostate_data['AGEP_A'].max()
bins = range(int(min_age), int(max_age) + 1, (int(max_age) - int(min_age)) // 20)

# Creating bins for age and custom labels
age_bins = pd.cut(final_prostate_data['AGEP_A'], bins=bins, right=False)
bin_labels = [f"{int(left)}-{int(right)-1}" for left, right in zip(bins[:-1], bins[1:])]

# Calculating frequency distribution
age_distribution = final_prostate_data.groupby(age_bins)['AGEP_A'].agg('count').reset_index(name='Count')

# Updating bin labels for plotting
age_distribution['AGEP_A'] = bin_labels

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='AGEP_A', y='Count', data=age_distribution, color='skyblue')
plt.xticks(rotation=45)
plt.title('Age Distribution of Patients')
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.show()
```

## The barplot shows the age distribution for the data. More patients ages 60 to 80. Also for this age distribution category, people prostate cancer is predominant in people between ages 84-86. Younger people are fewer in this data likewise.

```python
# Violin Plot
sns.violinplot(x=final_prostate_data['AGEP_A'], color='lightblue')
plt.title('Age Distribution of Patients - Violin Plot')
plt.xlabel('Age')
plt.show()
```
